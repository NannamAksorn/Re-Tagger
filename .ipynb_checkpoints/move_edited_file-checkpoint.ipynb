```python
import os
import glob
import shutil
```

```python
edited = glob.glob('key/*/*edit.csv')
for file in edited:
    print('edited_' + file[:9],'|||||', file)
    try:
        shutil.copy(file, 'edited_' + file)
    except Exception:
        os.mkdir('edited_' + file[:9])
        shutil.copy(file, 'edited_' + file)
```
